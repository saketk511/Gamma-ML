```python
match['Venue']=match['Venue'].replace('Arun Jaitley Stadium, Delhi','Arun Jaitley Stadium') 
match['Venue']=match['Venue'].replace('Brabourne Stadium, Mumbai','Brabourne Stadium') 
match['Venue']=match['Venue'].replace('Dr DY Patil Sports Academy, Mumbai','Dr DY Patil Sports Academy') 
match['Venue']=match['Venue'].replace('Eden Gardens, Kolkata','Eden Gardens') 
match['Venue']=match['Venue'].replace('Feroz Shah Kotla','Arun Jaitley Stadium') 
match['Venue']=match['Venue'].replace('M.Chinnaswamy Stadium','M Chinnaswamy Stadium') 
match['Venue']=match['Venue'].replace('MA Chidambaram Stadium, Chepauk, Chennai','MA Chidambaram Stadium') 
match['Venue']=match['Venue'].replace('MA Chidambaram Stadium, Chepauk','MA Chidambaram Stadium') 
match['Venue']=match['Venue'].replace('Maharashtra Cricket Association Stadium, Pune','Maharashtra Cricket Association Stadium') 
match['Venue']=match['Venue'].replace('Punjab Cricket Association IS Bindra Stadium','Punjab Cricket Association Stadium') 
match['Venue']=match['Venue'].replace('Punjab Cricket Association IS Bindra Stadium, Mohali','Punjab Cricket Association Stadium') 
match['Venue']=match['Venue'].replace('Punjab Cricket Association Stadium, Mohali','Punjab Cricket Association Stadium') 
match['Venue']=match['Venue'].replace('Rajiv Gandhi International Stadium, Uppal','Rajiv Gandhi International Stadium') 
match['Venue']=match['Venue'].replace('Wankhede Stadium, Mumbai','Wankhede Stadium') 
match['Venue'].value_counts()
```

```python
teams =[
    'Sunrisers Hyderabad', 'Mumbai Indians','Royal Challengers Bangalore','Kolkata Knight Riders','Punjab Kings',
    'Chennai Super Kings', 'Rajasthan Royals', 'Delhi Capitals', 'Lucknow Super Giants', 'Gujarat Titans'
]
match['Team1']=match['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match['Team2']=match['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match['TossWinner']=match['TossWinner'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match['WinningTeam'] = match['WinningTeam'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

match['Team1']=match['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match['Team2']=match['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')
match['TossWinner']=match['TossWinner'].str.replace('Delhi Daredevils','Delhi Capitals')
match['WinningTeam'] = match['WinningTeam'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match['Team1']=match['Team1'].str.replace('Kings XI Punjab','Punjab Kings')
match['Team2']=match['Team2'].str.replace('Kings XI Punjab','Punjab Kings')
match['TossWinner']=match['TossWinner'].str.replace('Kings XI Punjab','Punjab Kings')
match['WinningTeam'] = match['WinningTeam'].str.replace('Kings XI Punjab', 'Punjab Kings')


match['Team1']=match['Team1'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
match['Team2']=match['Team2'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
match['TossWinner']=match['TossWinner'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
match['WinningTeam'] = match['WinningTeam'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')

match = match[match['Team1'].isin(teams)]
match = match[match['Team2'].isin(teams)]
match = match[match['TossWinner'].isin(teams)]
match = match[match['WinningTeam'].isin(teams)]
```

```python
total_score = deliveries.groupby(['ID', 'innings']).sum()['total_run'].reset_index()
total_score.head()
```

```python
total_score = total_score[total_score['innings']==1]
```

```python
total_score['target'] = total_score['total_run'] + 1
match_df = match.merge(total_score[['ID','target']], on='ID')
```

```python
match_df = match_df[match_df['method'].isna()]
```

```python
match_df.head()
```

```python
deliveries['BattingTeam'] = deliveries['BattingTeam'].str.replace('Kings XI Punjab', 'Punjab Kings')
deliveries['BattingTeam'] = deliveries['BattingTeam'].str.replace('Delhi Daredevils', 'Delhi Capitals')
deliveries['BattingTeam'] = deliveries['BattingTeam'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
deliveries['BattingTeam'] = deliveries['BattingTeam'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
deliveries  = deliveries[deliveries['BattingTeam'].isin(teams)]
```

```python
deliveries_df =match_df.merge(deliveries,on='ID')
deliveries_df.head()
```

```python
deliveries_df.columns
```

```python
deliveries_df=deliveries_df[deliveries_df['innings']==2]
```

```python
deliveries_df.head()
```

```python
deliveries_df['current_score'] = deliveries_df.groupby('ID')['total_run'].cumsum()
deliveries_df['runs_left'] = deliveries_df['target']-deliveries_df['current_score']+1
```

```python
deliveries_df['balls_left']=120-(deliveries_df['overs']*6+deliveries_df['ballnumber'])
```

```python
deliveries_df['wickets_left'] = 10 - deliveries_df.groupby('ID')['isWicketDelivery'].cumsum()
```

```python
deliveries_df['crr']=(deliveries_df['current_score']*6)/(120-deliveries_df['balls_left'])
```

```python
deliveries_df['rrr']=(deliveries_df['runs_left']*6)/deliveries_df['balls_left']
```

```python
deliveries_df = deliveries_df.reset_index()
```

```python
def resultfun(row):
    
    return 1 if row['BattingTeam'] == row['WinningTeam'] else 0
```

```python
deliveries_df['result'] = deliveries_df.apply(resultfun,axis=1)
```

```python
index1 = deliveries_df[deliveries_df['Team2']==deliveries_df['BattingTeam']]['Team1'].index
index2 = deliveries_df[deliveries_df['Team1']==deliveries_df['BattingTeam']]['Team2'].index
deliveries_df.loc[index1, 'BowlingTeam'] = deliveries_df.loc[index1, 'Team1']
deliveries_df.loc[index2, 'BowlingTeam'] = deliveries_df.loc[index2, 'Team2']
deliveries_df.head()
```

```python
final_df = deliveries_df[['BattingTeam', 'BowlingTeam', 'City','runs_left',
                        'balls_left','wickets_left','crr',
                        'rrr', 'target', 'result']]

final_df.head()
```

```python
final_df.shape
```

```python
final_df.isnull().sum()
```

```python
# dropping of null values


final_df = final_df.dropna()

final_df.isnull().sum()
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df.shape
```

```python
data = final_df.copy()

test = data['result']

train = data.drop(['result'],axis = 1)

train.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(train,test,test_size=0.2,random_state=1)

X_train.shape,X_test.shape
```

```python
X_train.columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn import metrics
```

```python
# batting team,bowling team and city are categorical columns
# they will be converted to numeric using onehot encoder

cf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BowlingTeam','City'])
],remainder='passthrough')
```

```python
# creating the pipeline


pipe1 = Pipeline(steps=[
    ('step1',cf),
    ('step2',LogisticRegression(solver='liblinear'))
])

# fitting the training data

pipe1.fit(X_train,y_train)
```

```python
y_pred = pipe1.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```

```python
pipe1.predict_proba(X_test)[10]
```

```python
# Selecting Random Forest Classifier

pipe2 = Pipeline(steps=[
    ('step1',cf),
    ('step2',RandomForestClassifier())
])

pipe2.fit(X_train,y_train)
print(metrics.accuracy_score(y_test,pipe2.predict(X_test)))
```

```python
# oth index is loose prob and 1st index is win prob

pipe2.predict_proba(X_test)[10]
```
